```python
import pandas as pd

df1 = pd.read_csv("steam_games.xls")
df1 = df1.drop_duplicates(subset=["name"])

df2 = pd.read_csv("rawg_games_all.csv")
df2 = df2.drop_duplicates(subset=["name"])

merged = df1.merge(df2, on='name')
top1000 = merged.sort_values('rating', ascending=False).head(1000)
top1000.to_csv('top1000.csv', index=False)
```
