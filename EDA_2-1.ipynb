# EDA


## Первый df1

В датасете 20к строк с признаками игр, которые собраны парсингом с сайта: https://store.steampowered.com/search?term=

- name - название игры
- description - описание
- price - цена
- reviews - оценка словесная (Very Positive, Mostly Positive и другие)
- total_reviews - всего отзывов
- total_positive - сколько положительных отзывов
- total_negative - сколько отрицательных
- system_requirements - системные требования
- tags - теги
- url - ссылка



```python
import pandas as pd

df1 = pd.read_csv("steam_games_20к.xls")
print(df1.shape)
df1.head()
```

Удалим Nan значения и дублиакты:

```python
df1 = df1.dropna(subset=["name"])
df1 = df1.drop_duplicates(subset=["name"])
print(df1.shape)
df1.head()
```

## Второй df2

В датасете 20к строк с признаками игр, собранных с сайта: https://rawg.io/

- id игры
- name игры
- released - дата выпуска игры
- rating игры
- ratings_count	- количество отзывов
- metacritic - агрегированный рейтинг по шкале 0–100, рассчитанный как взвешенное среднее оценок критиков с сайта Metacritic
- playtime - количество часов прохождения
- genres - жанры, например: Strategy, Simulation
- platforms	- устройства, на которых можно играть
- screenshots_count	- количество скринов на сайте про эту игру
- achievements_count - количество достижений (ачивок), которые предусмотрены для этой игры

```python
df2 = pd.read_csv("rawg_games_20к.csv")
print(df2.shape)
df2.head()
```

Удалим Nan значения и дубликаты:

```python
df2 = df2.dropna(subset=["name"])
df2 = df2.drop_duplicates(subset=["name"])
print(df2.shape)
df2.head()
```

## Merge по name

```python
df_merged = df1.merge(df2, on='name', how='inner')
print(df_merged.shape)
df_merged.head()
```

# Обработка признаков


```python
df = df_merged.copy()
df.isna().sum()
```

```python
df.value_counts(["reviews"])
```

Топ оценок

1. Overwhelmingly Positive (примерно 97 % положительно)
2. Very Positive (примерно 86 % положительно)
3. Mostly Positive (примерно 71 % положительно)
4. Mixed (примерно 67 % положительно)
5. Mostly Negative (примерно 29 % положительно)

```python
rating_map = {
    "Mostly Negative": 1,
    "Mixed": 2,
    "Mostly Positive": 3,
    "Very Positive": 4,
    "Overwhelmingly Positive": 5}

df["reviews_score"] = df["reviews"].map(rating_map)
df.head()
```

```python
df = df.drop(columns=["reviews"]) # дублирует инфу из нового признака reviews_score
df = df.drop(columns=["url"]) # не показательный
df = df.drop(columns=['id']) # не показательный
df.head()
```

Обработка цены, оставляем просто число:

```python
df["price"] = df["price"].str.lower()
df.value_counts("price")
```

```python
price = df["price"].str.strip()

price = price.str.replace("usd", "", regex=False)
price = price.str.replace("$", "", regex=False)
price = price.str.strip()

df["price_usd"] = pd.to_numeric(price, errors="coerce")
df["price_usd"] = df["price_usd"].fillna(0)

df = df.drop(columns=["price"])
df.head()
```

- Удалим количество негативных и позитивных признаков, чтобы не было не нужных зависимостей с общим количеством
- Оставим только процент позитивных

```python
df["positive_ratio"] = df["total_positive"] / df["total_reviews"] * 100
df = df.drop(columns=["total_positive", "total_negative"])

df.head()
```

```python
df.value_counts(["achievements_count"])
```

- по идее библиотеки achievements_count должен был отражать кол-во достижений, которые можно получить в игре, но видимо на деле данные лажовые
- system_requirements дублирует platforms и сложно интерпретируется, а так же разного не структурированного формата, поэтому так же удалим

```python
df = df.drop(columns=["achievements_count", "system_requirements"])
```

Обработка признака даты релиза и создание признака год выпуска

```python
df["released"] = pd.to_datetime(df["released"], errors="coerce")
df["release_year"] = df["released"].dt.year
```

Оставим только количество слов в строке описания description, а сам текстовый признак описания удалим

```python
df['description_word_count'] = df['description'].str.split().str.len()
df = df.drop(columns=['description'])
```

Aналогично с tags

```python
df['count_tags'] = df['tags'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
```

## Обработка признака жанры

Далее идет обработка признака genres:

- добавляем столбец, в котором считается количество жарнов для каждой иргы
- добавляем 3 столбца со значениями первого, второго и третьего жанров в массиве в столбце "genres" для каждого элемента
- самые редко встречающиеся значения (меньше 40 от общего количества) записаны в один жанр (others)
- one-hot кодирование для этих признаков топ3 жанра по всем возможным жанрам
- изначальный столбец genres удаляем

```python
df['count_genres'] = df['genres'].apply(lambda x: len(x.split(', ')) if isinstance(x, str) else 0)
df['count_genres'].describe()
```

```python
split_genres = df['genres'].str.split(', ', expand=True)

split_genres = split_genres.apply(lambda x: x.str.strip() if isinstance(x, str) else x)

df['first_genre'] = split_genres[0].fillna('0')
df['second_genre'] = split_genres[1].fillna('0')
df['third_genre'] = split_genres[2].fillna('0')

df[['genres', 'first_genre', 'second_genre', 'third_genre']].head()
```

```python
print(df['first_genre'].nunique(), df['second_genre'].nunique(), df['third_genre'].nunique())
```

```python
first_genre_counts = df['first_genre'].value_counts()
second_genre_counts = df['second_genre'].value_counts()
third_genre_counts = df['third_genre'].value_counts()

small_first_genres = first_genre_counts[first_genre_counts < 40].index
small_second_genres = second_genre_counts[second_genre_counts < 40].index
small_third_genres = third_genre_counts[third_genre_counts < 40].index

df['first_genre'] = df['first_genre'].apply(lambda x: 'Other' if x in small_first_genres else x)
df['second_genre'] = df['second_genre'].apply(lambda x: 'Other' if x in small_second_genres else x)
df['third_genre'] = df['third_genre'].apply(lambda x: 'Other' if x in small_third_genres else x)

df[['first_genre', 'second_genre', 'third_genre']].head()
```

```python
df['first_genre'].value_counts()
```

```python
df['second_genre'].value_counts()
```

```python
df['third_genre'].value_counts()
```

```python
ohe_df = pd.get_dummies(df[['first_genre', 'second_genre', 'third_genre']].stack()).groupby(level=0).max()
ohe_df = ohe_df.astype(int)
df = pd.concat([df, ohe_df], axis=1)
df.drop(columns=['0']) #это были пропуски в жанрах
df.head()
```

## Анализ признака platforms

При обработке признака platforms:
- добавили признак количества устройств, на которых поддерживается каждая игра (platform_count)
- объединили все версии PlayStation, Xbox и других платформ

```python
df['platform_count'] = df['platforms'].apply(lambda x: len(x.split(',')))
df['platform_count'].describe()
```

```python
platform_count = {}

for i in df['platforms']:
    for platform in i.split(','):
        platform = platform.strip()
        if platform in platform_count:
            platform_count[platform] += 1
        else:
            platform_count[platform] = 1

platform_count
```

```python
platform_groups = {
    'PlayStation': ['PlayStation', 'PlayStation 5', 'PlayStation 4', 'PlayStation 3', 'PS Vita', 'PlayStation 2'],
    'Xbox': ['Xbox', 'Xbox One', 'Xbox Series S/X', 'Xbox 360'],
    'Nintendo': ['Nintendo Switch', 'Nintendo DS', 'Nintendo 64', 'GameCube', 'Nintendo 3DS'],
    'SEGA': ['SEGA Saturn', 'SEGA CD', 'Genesis', 'SEGA Master System'],
    'Atari': ['Atari ST', 'Atari 8-bit', 'Atari 5200', 'Atari 2600', 'Atari Lynx'],
    'Mac': ['macOS', 'Classic Macintosh'],
    'PC': ['PC'],
    'Linux': ['Linux'],
    'Android': ['Android'],
    'iOS': ['iOS']}
```

```python
mapping = {}
for group, platforms in platform_groups.items():
    group_count = 0
    for platform in platforms:
        group_count += platform_count.get(platform, 0)
    mapping[group] = group_count

mapping
```

```python
#удалим наименее популярные платформы
del platform_groups['SEGA']
del platform_groups['Atari']
```

```python
for platform_group, platforms in platform_groups.items():
    df[platform_group] = df['platforms'].apply(lambda x: 1 if any(platform in x.split(',') for platform in platforms) else 0)

df = df.drop(columns=['platforms'])
df.head()
```

```python
def count_platform_types(row):
    types = 0

    if row['PlayStation'] or row['Xbox'] or row['Nintendo']:
        types += 1  # Консоли

    if row['Mac'] or row['PC'] or row['Linux']:
        types += 1  # Компьютеры

    if row['Android'] or row['iOS']:
        types += 1  # Телефоны
    return types

df['cross_platform'] = df.apply(count_platform_types, axis=1)
df
```

# Регрессия


У нас таргет metacritic (отзывы квалифицированных критиков), попробуем его предсказать для пропущенных объектов

```python
df_num = df.select_dtypes(include='number')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

df_known = df_num.dropna(subset=["metacritic"])
df_unknown = df_num[df_num["metacritic"].isna()]

numeric_cols = df_num.select_dtypes(include="number").columns
feature_cols = numeric_cols.drop("metacritic")

X_train = df_known[feature_cols]
y_train = df_known["metacritic"]
X_pred = df_unknown[feature_cols]

imputer = SimpleImputer(strategy="mean")
scaler = StandardScaler()

X_train_imp = imputer.fit_transform(X_train)
X_pred_imp = imputer.transform(X_pred)

X_train_scaled = scaler.fit_transform(X_train_imp)
X_pred_scaled = scaler.transform(X_pred_imp)

linreg = LinearRegression().fit(X_train_scaled, y_train)
metacritic_pred = linreg.predict(X_pred_scaled)

df_filled = df_num.copy()
df_filled.loc[df_filled["metacritic"].isna(), "metacritic"] = metacritic_pred

coef_table = pd.DataFrame({"feature": feature_cols, "coef": linreg.coef_})

coef_table = coef_table.reindex(
    coef_table["coef"].abs().sort_values(ascending=False).index
)

print("Топ10 признаков по важности:")
print(coef_table.head(10))
```
